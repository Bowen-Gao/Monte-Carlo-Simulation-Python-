### Part (a)

```python
x = []
for i in range(10000):
     x.append(random.uniform(-1, 1))
plt.hist(x, bins = 50, normed = True)
plt.show() 
```

### Part (b)

```python
X = []
for i in range(10000):
     X.append(random.uniform(-1, 1))
        
Y = []
for i in range(10000):
     Y.append(random.uniform(-1, 1))
        
Z = []
for i in range(10000):
    if (X[i]**2 + Y[i]**2) <= 1:
        Z.append(1)
    else:
        Z.append(0)   
```

```python
plt.scatter(X, Y, c = Z, s = 2.0, cmap = plt.cm.Paired)
plt.show()        
```

### Part (c)

```python
def rel_error(n):
    X = []
    for i in range(n):
        X.append(random.uniform(-1, 1))
    Y = []
    for i in range(n):
        Y.append(random.uniform(-1, 1))
    Z = []
    for i in range(n):
        if (X[i]**2 + Y[i]**2) <= 1:
            Z.append(1)
        else:
            Z.append(0)
    estimated_pi = 4*float(sum(Z)) / float(n)
    rel_err = abs(estimated_pi - math.pi) / math.pi
    return rel_err

x = []
for i in range(1000):
    x.append(rel_error(i + 1))
plt.scatter(range(1, 1001), x, 15, 'g')

y = []
for i in range(1000):
    y.append(1 / math.sqrt(i + 1))
plt.plot(range(1, 1001), y, 'r')
plt.show()
```
